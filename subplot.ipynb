```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x= np.arange(1,6)
```

```python
x
```

```python
y=np.arange(1,11,2)
```

```python
y
```

```python
plt.plot(x,y,"blue")
plt.show()
```

```python
plt.subplot(2,3,1)
plt.plot(x,y,"red")
plt.subplot(2,3,2)
plt.plot(y,x,"blue")
plt.subplot(2,3,3)
plt.plot(x*x,y*y,"green")
plt.subplot(2,3,4)
plt.plot(x*y,y,"orange")
plt.show()
```

```python
fig =plt.figure()

axes = fig.add_axes([0.1,0.2,0.4,0.7]) #[x,y,xlong,ylong]

axes.plot(x,y)

axes.set_xlabel("outer x")
axes.set_ylabel("outer y")
axes.set_title("graph")
plt.show()
```
